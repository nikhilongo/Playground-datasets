```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
[df[col].value_counts() for col in df]
```

```python
df.drop(columns=['PassengerId','Cabin','Name'],inplace=True)
```

```python
df.isnull().sum().reset_index()
```

```python
df[df['CryoSleep'].isnull()]
```

```python
col=[col for col in df if df[col].dtypes=='float64']
for i in col:
    df[i].fillna(df[i].mean(),inplace=True)
```

```python
col1=[col1 for col1 in df if df[col1].dtypes=='object']
for i in col1:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df1 = pd.get_dummies(df,columns=col1,drop_first=True)
```

```python
y = df['Transported']
```

```python
x = df1.drop(columns=['Transported'])
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
x_flatten = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_flatten,y,test_size=0.3)
```

```python
x_train.shape
```

```python
import tensorflow as tf
import tensorflow.keras as k
from keras.models import Sequential
```

```python
model = Sequential([
    k.layers.Dense(24,input_shape=(12,),activation='relu'),
    k.layers.Dense(12,activation='relu'),
    k.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_flatten,y,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
predict = model.predict(x_test)
```

```python
predict[2]
```

```python
predicted = []
for i in predict:
  if(i>0.5):
    predicted.append(1)
  else:
    predicted.append(0)
```

```python
tf.math.confusion_matrix(y_test,predicted)
```
